# Mtplotlib with Seaborn

```python
import matplotlib.pyplot as plt
```

```python
def plot_lines(x1, y1, x2, y2):
    plt.figure()
    plt.plot(x1, y1, label='Mock Data 1')
    plt.plot(x2, y2, label='Mock Data 1')
    plt.title('Some mock data for demo')
    plt.xlabel('x axis')
    plt.ylabel('Mock Data')
    plt.legend()
    plt.show()
plot_lines([2,4,6,8,10,12], [3,5,9,6,12, 10], [2,4,6,8,10,12], [2,7,9,15,17,4])
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
sns.relplot(x='year', y='passengers', data=flights)
```

```python
combined_years = flights.groupby('year', as_index=False).sum()
```

```python
combined_years
```

```python
sns.relplot(x='year', y='passengers', data=combined_years)
```

## Train a model using Seaborn

```python
sns.lmplot(x='year', y='passengers', data=combined_years)
```

### BarPlot

```python
sns.barplot(x='year', y='passengers', data=combined_years) # Combined Data
```
